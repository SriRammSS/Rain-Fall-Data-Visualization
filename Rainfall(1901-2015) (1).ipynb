```python
import numpy as np
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
pd.set_option('display.max_row',None)
a=pd.read_csv('rainfall_area-wt_India_1901-2015.csv')
plt.rcParams['figure.figsize']=(20,10)
fg, ax = plt.subplots()
ax.plot(a['YEAR'],a['ANNUAL'],marker='.',markersize=15,markerfacecolor='#8989bc',alpha=0.7,markeredgecolor='#6060aa',markeredgewidth=2 ,label='Average Year',color='#8989bc')
ax.set_xlabel('Years',fontsize=20)
ax.set_ylabel('Rain Fall Density in mm',fontsize=20)
ax.set_title('Rainfall Variation of India (1901-2015)',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

```python
x='YEAR'
y=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
t=np.arange(115)
for i in y:
    plt.rcParams['figure.figsize']=(10,5)
    fg, ax = plt.subplots()
    ax.scatter(a[x],a[i],c=t,cmap='viridis',alpha=0.8,s=a[i] )
    ax.set_xlabel(x)
    ax.set_ylabel('Avg Rainfall in mm')
    ax.set_title("Rainfall in {} ".format(i))

    
    plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
fg, ax = plt.subplots()
plt.rcParams['figure.figsize'] = (10,10)
ax.scatter(a['YEAR'],a['Jan-Feb'],color='green',label='Jan-Feb',alpha=0.8,s=a['Jan-Feb'])
ax.scatter(a['YEAR'],a['Mar-May'],color='red',label='Mar-May',alpha=0.8,s=a['Mar-May'])
ax.scatter(a['YEAR'],a['Jun-Sep'],color='orange',label='Jun-Sep',alpha=0.8,s=a['Jun-Sep'])
ax.scatter(a['YEAR'],a['Oct-Dec'],color='blue',label='Oct-Dec',alpha=0.8,s=a['Oct-Dec'])
ax.set_xlabel('YEAR',fontsize=20)
ax.set_ylabel('Averagen Rainfall Density in mm',fontsize=20)
ax.set_title('Average RainFall(1901-2015)',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc='center left',fontsize=20)
plt.show()
```
